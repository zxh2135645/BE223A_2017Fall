## Identify the unique values and calculate their frequency

```python
def get_uniq_value_and_freq(data):
    fields = data.keys()
    #print(fields)
    total = data.shape[0]
    unique_value_list = []
    freq_list = []
    array_list = []
    counts_list = []

    for i, f in enumerate(fields):
        array = data[f]
        array = array.dropna()
        unique_values, counts = np.unique(array, return_counts=True)
        #print(unique_values, counts)
        array_size = np.size(array)
        nan_count = total - array_size
        freq = np.divide(counts, total)
        unique_value_list.append(unique_values)
        freq_list.append(freq)
        array_list.append(array)
        counts_list.append(counts)
        print("For column {0} {1}, the present unique values are: {2}. "
              "\nAnd the frequency for them are {3} respectively\n".format(i+1, f, unique_values, freq))
        print("##################################################################################")
    
    unique_dict = dict(zip(fields, unique_value_list))
    freq_dict = dict(zip(fields, freq_list))
    return array_list, counts_list, unique_dict, freq_dict

array_list, counts_list, unique_dict, freq_dict = get_uniq_value_and_freq(data)
#print(unique_dict)
```

* In order to have a better idea of each unique value in one column, the histogram is log scaled.

```python
def plot_numerical_hist(array, f, unique_values):
    bin_num = len(unique_values)
    if len(unique_values) == 4:
        bin_num = 8
    plt.hist(array, bins=bin_num, range=[np.min(unique_values), np.max(unique_values)], log=True, facecolor='blue', label=f, alpha=0.5)
    plt.title('Histogram of {0}'.format(f), fontdict = {'fontsize':9.5})
    plt.xlabel('Values', fontsize=8)
    plt.ylabel('Counts', fontsize=8)
    #plt.legend()
    plt.show()

fields = data.keys()
#print(array_list[0])
plt.figure()
plt.style.use('ggplot')
#plt.subplot()
#print(type(unique_dict['Region'][0]) is str)
count = 1

for i, f in enumerate(fields[1:]): # Excluding Column 1: ID
    unique_values = unique_dict[f]
    #print(unique_values)
    if type(unique_values[0]) is not str:
        num = 220 + count
        plt.subplot(num)
        array = array_list[i+1]
        plot_numerical_hist(array, f, unique_values)
        count = count + 1
        
```

```python
def get_nan_freq(data):
    fields = data.keys()
    total = data.shape[0]
    nan_freq_list = []
    

    for i, f in enumerate(fields):
        array = data[f]
        array = array.dropna()
        array_size = np.size(array)
        nan_count = total - array_size
        #print(nan_count)
        nan_freq = np.divide(nan_count, total)
        percentage_nan = 100 * nan_freq
        str_percent_nan = str(round(percentage_nan, 1)) + '%'
        #print(nan_freq)
        nan_freq_list.append(nan_freq)
        print("For column {0} {1}, the frequency of values are missing is: {2}. "
              "And the percentage for them are {3} respectively\n".format(i, f, nan_freq, str_percent_nan))
        print("##################################################################################")
    
    nan_dict = dict(zip(fields, nan_freq_list))
    return nan_dict

nan_dict = get_nan_freq(data)
#print(nan_dict)
```

## Part 2. Evaluating the performance of the model
#### Data Manipulation

```python
# Read the Column Classifier Result
classifier_result = data['Classifier Result']

threshold = 0.3
classifier_result[classifier_result >= threshold] = 1
classifier_result[classifier_result < threshold] = 0
classifier_result = classifier_result.astype('int64')
ground_truth = data['Ground Truth']
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import itertools
tn, fp, fn, tp = confusion_matrix(ground_truth, classifier_result).ravel()
cm = confusion_matrix(ground_truth, classifier_result)
classes = ['0', '1']

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

plt.figure()
plot_confusion_matrix(cm, classes, title='Confusion matrix, without normalization')
plt.show()
```

#### Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ground_truth, classifier_result)
print("The accuracy score is {0:.3f}".format(accuracy))
```

#### Recall

```python
from sklearn.metrics import recall_score
recall = recall_score(ground_truth, classifier_result, average='binary')
print("The recall score is {0:.3f}".format(recall))
```

#### Precision

```python
from sklearn.metrics import precision_score
precision = precision_score(ground_truth, classifier_result, average='binary')
print("The precision score is: {0:.3f}".format(precision))
```

#### F1 Score

```python
from sklearn.metrics import f1_score
f1 = f1_score(ground_truth, classifier_result, average='binary')
print("The f1 score is: {0:.3f}".format(f1))
```
