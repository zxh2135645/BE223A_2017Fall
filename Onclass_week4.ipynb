```python
import matplotlib.pylab as plt
import numpy as np

img = plt.imread('../boneage-training-dataset/1459.png')
print(img)
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.hist(img.ravel(), bins=256, range=(0.0, 1.0))
plt.show()
```

```python
plt.imshow(img, cmap='gray', clim=(0.5, 1))
plt.show()
```

```python
def simple_thrshold(im, threshold):
    idx = im[:,:] < threshold
    im[idx] = 0.0
    return im

plt.imshow(simple_thrshold(img, 0.5), cmap='gray')
plt.show()
```

```python
# scaling
factor = 2
[W, H] = img.shape
newW = int(W*factor)
newH = int(H*factor)

newImage = np.empty([newW, newH])
for col in range(newW):
    for row in range(newH):
        pixel = img[int(col/factor), int(row/factor)]
        newImage[col, row] = pixel
        
plt.imshow(newImage, cmap='gray')
plt.show()
```
